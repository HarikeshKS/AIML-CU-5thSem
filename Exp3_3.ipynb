```python
# [1] Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits

# [2] Loading Iris dataset and creating a DataFrame
digits = load_digits()
df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
df['target'] = digits.target
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# [3] Displaying first few rows and summary statistics
print(df.head())
print(df.describe())

# [4] Data visualization using Se33aborn and Matplotlib
sns.pairplot(df, hue='species')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()

plt.figure(figsize=(10, 6))
for species in df['species'].unique():
    sns.distplot(df[df['species'] == species]['petal width (cm)'], label=species)

plt.xlabel('Petal width (cm)')
plt.legend()
plt.show()

correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.show()
```
